# PRINT THE FOLLOWING PATTERN SHOWN ABOVE:


```python
for i in range(5,0,-1):
    for j in range(1,6-i):
        print(" ",end = "")
    for k in range(1,i+1):
        print("*",end="")
    print('\n')
```
